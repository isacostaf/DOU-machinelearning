```python
!pip install PyPDF2 scikit-learn nltk

import os
from PyPDF2 import PdfReader
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.svm import LinearSVC
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

# função para extrair texto
def extrair_texto_pdf(caminho):
    texto = ""
    try:
        reader = PdfReader(caminho)
        for page in reader.pages:
            texto += page.extract_text() or ""
    except:
        pass
    return texto

# carregar dados
texts = []
labels = []

base_path = "/Users/isacosta/Documents/M. Defesa/Representacoes-ML"

for label in ["representacao", "nao_representacao"]:
    pasta = os.path.join(base_path, label)
    
    for arquivo in os.listdir(pasta):
        if arquivo.endswith(".pdf"):
            caminho = os.path.join(pasta, arquivo)
            texto = extrair_texto_pdf(caminho)
            
            texts.append(texto)
            labels.append(label)

# divisão treino/teste
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=5, random_state=42, stratify=labels
)

# stopwords
stopwords_pt = stopwords.words('portuguese')

# vetor
vectorizer = TfidfVectorizer(
    stop_words=stopwords_pt,
    max_features=1000,
    ngram_range=(1,2)
)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# modelo
model = LinearSVC(class_weight='balanced')
model.fit(X_train_vec, y_train)

# previsão
y_pred = model.predict(X_test_vec)

print(classification_report(y_test, y_pred))
```
